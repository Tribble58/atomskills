```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

font = 16
sns.set(font_scale=1.4)
```

```python
data_names = ['SOP1.dat', 'SOP2.dat', 'SOP3.dat']
all_data = {}
for file in data_names:
    path_to_data = 'C:/Users/Vladlen/Documents/Python/AtomSkills/data/' + file
    f = open(path_to_data).read()
    all_data[file] = f.split('\n')
```

```python
len(all_data['SOP1.dat']), len(all_data['SOP2.dat']), len(all_data['SOP3.dat'])
```

Удаление технической информации

```python
for data in data_names:
    all_data[data] = all_data[data][1:-1]
```

```python
all_data
```

```python
df = pd.DataFrame(columns=['filename', 'index'] + [f'schema_{i}' for i in range(1, 17)])
for data in data_names:
    for values in all_data[data]:
        values_list = values.split(' ')
        values_list = list(map(int, values_list))
        df.loc[len(df), :] = [data] + values_list
df
```

```python
df.info()
```

Типы нуждаются в преобразовании:

```python
numeric_columns = [col for col in df.columns if col not in ['filename', 'index']]
df[numeric_columns] = df[numeric_columns].astype('int')
```

```python
df.describe()
```

### Визуализация

```python
def draw_plots(filename):
    fig, axs = plt.subplots(16, 1, figsize=(20, 60))
    for k in range(16):
        axs[k].plot(df[df.filename == filename]['index'], df[df.filename == filename][f'schema_{k+1}'])
        axs[k].set_xlabel(f'schema_{k+1}', fontsize=font)
        axs[k].set_title(filename, fontsize=font)
    plt.tight_layout()
```

```python
def draw_2_plots(schema_1, schema_2):
    plt.figure(figsize=(20, 5))
    plt.plot(df[df.filename == 'SOP1.dat']['index'], df[df.filename == 'SOP1.dat'][schema_1], label=schema_1)
    plt.plot(df[df.filename == 'SOP1.dat']['index'], df[df.filename == 'SOP1.dat'][schema_2], label=schema_2)
    plt.legend(fontsize=font)
```

```python
draw_plots('SOP1.dat')
```

### Корреляция

```python
corr = df[df.filename == 'SOP1.dat'][numeric_columns].corr()
plt.figure(figsize=(20, 10))
sns.heatmap(corr, cmap='Blues', annot=True, fmt='.2f')
```

Сильная линейная зависимость между 2 и 12, 4 и 10 схемами, средняя - между 8 и 16 и 13 и 14
